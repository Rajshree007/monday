```python
import gym

env = gym.make('Blackjack-v1')
def run_blackjack_game():
    state = env.reset()

    while True:
        env.render()

        action = env.action_space.sample()
        print(action)

        next_state, reward, done, info = env.step(action)
        print(next_state)
        print(reward)

        if done:
            env.render()
            print(f"End game! Reward: {reward}")
            if reward > 0:
                print("You won! :)")
            elif reward < 0:
                print("You lost! :(")
            else:
                print("It's a draw!")
            break

for i_episode in range(10):
    print(f"Episode {i_episode + 1}:")
    run_blackjack_game()
    print("\n" + "="*50 + "\n")
```

```python
import gym


env = gym.make('Blackjack-v1')
def simple_blackjack_agent(state):
    player_score, dealer_card, usable_ace = state

    if player_score < 17:
        return 1
    else:
        return 0

num_episodes = 10

for episode in range(num_episodes):
    state = env.reset()
    done = False
    total_reward = 0

    while not done:
        action = simple_blackjack_agent(state)
        next_state, reward, done, _ = env.step(action)


        env.render()

        state = next_state
        total_reward += reward

    print(f"Episode {episode + 1}, Total Reward: {total_reward}")
```
